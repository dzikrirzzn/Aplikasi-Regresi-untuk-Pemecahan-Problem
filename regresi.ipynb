```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

# Memuat dataset dari file yang diunggah
file_path = '/content/Student_Performance.csv'  # Sesuaikan dengan nama file Anda
data = pd.read_csv(file_path)

# Mengekstrak kolom yang relevan
NL = data['Sample Question Papers Practiced']
NT = data['Performance Index']

# Mengubah data ke dalam bentuk yang sesuai untuk regresi
NL = NL.values.reshape(-1, 1)
NT = NT.values.reshape(-1, 1)

# Membuat dan menyesuaikan model regresi linear
linear_model = LinearRegression()
linear_model.fit(NL, NT)

# Prediksi
NT_pred_linear = linear_model.predict(NL)

# Plot hasil
plt.scatter(NL, NT, color='blue')
plt.plot(NL, NT_pred_linear, color='red')
plt.title('Linear Regression')
plt.xlabel('Sample Question Papers Practiced')
plt.ylabel('Performance Index')
plt.show()

# Menghitung galat RMS
rms_linear = np.sqrt(mean_squared_error(NT, NT_pred_linear))
print(f"RMS Linear Model: {rms_linear}")
from scipy.optimize import curve_fit

# Fungsi model pangkat sederhana
def power_model(x, a, b):
    return a * np.power(x, b)

# Menyesuaikan model
params, covariance = curve_fit(power_model, NL.flatten(), NT.flatten())
a, b = params

# Prediksi
NT_pred_power = power_model(NL, a, b)

# Plot hasil
plt.scatter(NL, NT, color='blue')
plt.plot(NL, NT_pred_power, color='red')
plt.title('Power Regression')
plt.xlabel('Sample Question Papers Practiced')
plt.ylabel('Performance Index')
plt.show()

# Menghitung galat RMS
rms_power = np.sqrt(mean_squared_error(NT, NT_pred_power))
print(f"RMS Power Model: {rms_power}")
```
